### Setting up selenium here with all required attributes

```python
# install chromium, its driver, and selenium
!apt update
!apt install chromium-chromedriver
!pip install selenium
# set options to be headless, ..
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
# The files will be downloaded onto default directories of colab and input.xlsx file will be imported as a dataframe here
!pip install -q xlrd
import pandas as pd
df = pd.read_excel('/content/20211030 Test Assignment/Input.xlsx')
df
```

###Using For loop to go through each url, extract title and content
###Save to a file with url id as file name 

```python
#Creating for loop to take each url from dataframe and pass it to selenium to extract title and content
#Loop through each row
#This code stores title as first line and the content after a new line i.e., on the Second line
!mkdir takeout #This is the directory where text files will be stored
from selenium.webdriver.common.by import By
wd = webdriver.Chrome(options=options)
for i in range(df.shape[0]):
  url = df.at[i,"URL"]
  wd.get(str(url))

  title = wd.title
  s = ''

  elements = wd.find_elements(By.TAG_NAME, 'p')

  for e in elements:
    s += e.text
  
  #start writing contents into text file and save them
  text_file = open("/content/takeout/" + str(i + 1) + ".txt", "w")

  text_file.write(title + "\n" + s)

  text_file.close()
  print("File Created:" + str(i+1))
```
